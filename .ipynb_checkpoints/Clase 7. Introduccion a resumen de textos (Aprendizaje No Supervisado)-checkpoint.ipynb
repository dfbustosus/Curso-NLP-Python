# Aprendizaje no supervisado 

Que es el aprendizaje no supervisado?

El aprendizaje no supervisado es un tipo de aprendizaje automtico en donde modelo se entrena con datos no etiquetados, sin ningn conocimiento previo del resultado correcto. El objetivo del aprendizaje no supervisado es encontrar patrones o relaciones en los datos, como agrupar elementos similares, detectar anomalas o reducir la dimensionalidad de los datos. 

Estas son algunas de las cosas que podemos hacer con el aprendizaje no supervisado de texto:

- Extraiga las frases influyentes clave de los documentos
- Extraer diversos conceptos o temas presentes en el documentos
- Resumir los documentos para proporcionar una esencia que retenga la partes importantes de todo el corpus

El resumen de texto y la extraccin de informacin se ocupan de tratar de extraer conceptos y temas importantes clave de un gran corpus de texto, esencialmente reducindolo. Ahora para que lo necesitamos?.

El concepto de sobrecarga de informacin es una de las principales razones detrs de la demanda de resmenes de texto. Desde que los medios impresos y verbales cobraron importancia, ha habido una gran cantidad de libros, artculos, audio y video disponibles.



Esto comenz en el siglo III o IV a. C., cuando la gente se refera a una gran cantidad de libros, ya que pareca no haber fin a la produccin de libros. El Renacimiento nos dio la invencin de la imprenta por Gutenberg alrededor de 1440 d.C., lo que condujo a la produccin en masa de libros, manuscritos, artculos y folletos. Esto aument considerablemente la sobrecarga de informacin, y los acadmicos se quejaron de un exceso de informacin, que se estaba volviendo extremadamente difcil de consumir, procesar y administrar.



En el siglo XX, los avances en las computadoras y la tecnologa marcaron el comienzo de la era digital, que culmin con Internet. Internet abri toda una ventana de posibilidades para producir y consumir informacin con las redes sociales, los sitios web de noticias, el correo electrnico y las capacidades de mensajera instantnea. Esto, a su vez, ha llevado a un aumento explosivo en la cantidad de informacin y de informacin no deseada en forma de spam, estados no deseados y tweets, e incluso a bots que publican ms contenido no deseado en la Web.



La sobrecarga de informacin, entonces, es la presencia de datos o informacin en exceso, que los consumidores encuentran difcil de procesar para tomar decisiones bien informadas. La sobrecarga ocurre cuando la cantidad de informacin que ingresa al sistema comienza a exceder la capacidad de procesamiento del sistema. 

Las personas de altos cargos ejecutivos rara vez tienen tiempo para escuchar largas charlas o leer pginas y pginas de informacin importante. La idea de resumir y extraer informacin es tener una idea de los temas y temas clave importantes de grandes documentos de informacin y resumirlos en unas pocas lneas que se pueden leer, entender e interpretar fcilmente, facilitando as el proceso de tomar decisiones bien informadas.



Y que podemos hacer entonces, bueno en general tenemos tres posibilidades:

1. Keyphrase extraction (extraccion de palabras/frases clave)
2. Topic modeling (Modelado de topicos)
3. Automated document summarization (Resumenes automaticos de documentos)

Vamos a explicar cada uno

1. **La extraccin de frases clave** es quizs la ms simple de las tres tcnicas. Implica extraer palabras clave o frases de un documento o corpus de texto que capturen sus principales conceptos o temas. Se puede decir que esto es una forma simplista de modelado de temas. Es posible que hayas visto palabras clave o frases descritas en un trabajo de investigacin o incluso algn producto en una tienda en lnea que describa la entidad en unas pocas palabras o frases, capturando su idea o concepto principal.

2. **Topic Modeling** generalmente implica el uso de tcnicas de modelado estadstico y matemtico para extraer temas, temas o conceptos principales de un corpus de documentos (ms diverso sea el conjunto de documentos que tenga, ms temas o conceptos puedes generar). Hay modelos estadsticos probabilsticos, que utilizan tcnicas estadsticas especficas que incluyen la descomposicin de valores singulares y el LDA para descubrir estructuras semnticas latentes conectadas en datos de texto que generan temas y conceptos. Se utilizan ampliamente en anlisis de texto e incluso en bioinformtica.

3. **El resumen automatizado de documentos** es el proceso de utilizar un programa informtico o algoritmo basado en tcnicas estadsticas y de aprendizaje automtico para resumir un documento o corpus de documentos. Existe una amplia variedad de tcnicas para crear resmenes de documentos automatizados, incluidas varias tcnicas basadas en extraccin y abstraccin.

El concepto clave detrs de todos estos algoritmos es encontrar un subconjunto representativo del conjunto de datos original de modo que la esencia central del conjunto de datos desde los puntos de vista semntico y conceptual est contenida en este subconjunto. Esto es un concepto recurrente que se puede aplicar a imagenes y texto

# Conceptos que debes conocer antes de empezar

 
 
1. **Documentos**:  una entidad que contiene texto y metadatos. Un corpus generalmente consiste en una coleccin de documentos. Estos documentos pueden ser oraciones simples o prrafos completos de informacin textual. El corpus tokenizado se refiere a un corpus en el que cada documento se tokeniza o se divide en tokens, que generalmente son palabras.

2. **Normalizacin de texto** proceso de limpieza, normalizacin y estandarizacin de datos textuales con tcnicas como la eliminacin de smbolos y caracteres especiales, la eliminacin de etiquetas HTML extraas, la eliminacin de palabras vacas, la correccin ortogrfica, stemming y la lematizacin.

3. **Feature extraction** es un proceso mediante el cual extraemos caractersticas o atributos significativos de datos textuales sin procesar para introducirlos en un algoritmo estadstico o de aprendizaje automtico. Este proceso tambin se conoce como vectorizacin porque, por lo general, la transformacin final de este proceso son vectores numricos a partir de tokens de texto sin procesar. La razn es que los algoritmos convencionales funcionan con vectores numricos y no pueden funcionar directamente con datos de texto sin procesar. Tenemos distintos mecanismos de abstraccion como BoW, TFIDF, Word Embeddings (**No te preocupes mas adelante vamos a ver mas a detalle esto)

4. **Matriz de diseo** se refiere a un mapeo de una coleccin de documentos a caractersticas donde cada fila indica un documento y cada columna indica una caracterstica particular, generalmente una palabra o un conjunto de palabras. 

5. **Descomposicion SVD** La descomposicin en valores singulares (SVD) es una tcnica del lgebra lineal que se utiliza con bastante frecuencia en los algoritmos de resumen. SVD es el proceso de factorizacin de una matriz que es real o compleja. 



Los valores singulares de $\Sigma$ son particularmente importantes en los algoritmos de resumen ahora te preguntaras porque?. 

Usaremos SVD particularmente para la aproximacin de matrices de bajo rango donde aproximamos la matriz original M con una matriz M tal que esta nueva matriz es una versin truncada de la matriz original M con un rango k y puede ser representada por SVD como $M = U \Sigma V^*$ 

Donde $\Sigma$ es una versin truncada de la matriz $\Sigma$  original, que ahora consiste solo en los k valores singulares ms grandes, y los otros valores singulares estn representados por cero.
