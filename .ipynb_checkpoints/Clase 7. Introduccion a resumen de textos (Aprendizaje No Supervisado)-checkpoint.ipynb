# Aprendizaje no supervisado 

Que es el aprendizaje no supervisado?

El aprendizaje no supervisado es un tipo de aprendizaje automático en donde modelo se entrena con datos no etiquetados, sin ningún conocimiento previo del resultado correcto. El objetivo del aprendizaje no supervisado es encontrar patrones o relaciones en los datos, como agrupar elementos similares, detectar anomalías o reducir la dimensionalidad de los datos. 

Estas son algunas de las cosas que podemos hacer con el aprendizaje no supervisado de texto:

- Extraiga las frases influyentes clave de los documentos
- Extraer diversos conceptos o temas presentes en el documentos
- Resumir los documentos para proporcionar una esencia que retenga la partes importantes de todo el corpus

El resumen de texto y la extracción de información se ocupan de tratar de extraer conceptos y temas importantes clave de un gran corpus de texto, esencialmente reduciéndolo. Ahora para que lo necesitamos?.

El concepto de sobrecarga de información es una de las principales razones detrás de la demanda de resúmenes de texto. Desde que los medios impresos y verbales cobraron importancia, ha habido una gran cantidad de libros, artículos, audio y video disponibles.



Esto comenzó en el siglo III o IV a. C., cuando la gente se refería a una gran cantidad de libros, ya que parecía no haber fin a la producción de libros. El Renacimiento nos dio la invención de la imprenta por Gutenberg alrededor de 1440 d.C., lo que condujo a la producción en masa de libros, manuscritos, artículos y folletos. Esto aumentó considerablemente la sobrecarga de información, y los académicos se quejaron de un exceso de información, que se estaba volviendo extremadamente difícil de consumir, procesar y administrar.



En el siglo XX, los avances en las computadoras y la tecnología marcaron el comienzo de la era digital, que culminó con Internet. Internet abrió toda una ventana de posibilidades para producir y consumir información con las redes sociales, los sitios web de noticias, el correo electrónico y las capacidades de mensajería instantánea. Esto, a su vez, ha llevado a un aumento explosivo en la cantidad de información y de información no deseada en forma de spam, estados no deseados y tweets, e incluso a bots que publican más contenido no deseado en la Web.



La sobrecarga de información, entonces, es la presencia de datos o información en exceso, que los consumidores encuentran difícil de procesar para tomar decisiones bien informadas. La sobrecarga ocurre cuando la cantidad de información que ingresa al sistema comienza a exceder la capacidad de procesamiento del sistema. 

Las personas de altos cargos ejecutivos rara vez tienen tiempo para escuchar largas charlas o leer páginas y páginas de información importante. La idea de resumir y extraer información es tener una idea de los temas y temas clave importantes de grandes documentos de información y resumirlos en unas pocas líneas que se pueden leer, entender e interpretar fácilmente, facilitando así el proceso de tomar decisiones bien informadas.



Y que podemos hacer entonces, bueno en general tenemos tres posibilidades:

1. Keyphrase extraction (extraccion de palabras/frases clave)
2. Topic modeling (Modelado de topicos)
3. Automated document summarization (Resumenes automaticos de documentos)

Vamos a explicar cada uno

1. **La extracción de frases clave** es quizás la más simple de las tres técnicas. Implica extraer palabras clave o frases de un documento o corpus de texto que capturen sus principales conceptos o temas. Se puede decir que esto es una forma simplista de modelado de temas. Es posible que hayas visto palabras clave o frases descritas en un trabajo de investigación o incluso algún producto en una tienda en línea que describa la entidad en unas pocas palabras o frases, capturando su idea o concepto principal.

2. **Topic Modeling** generalmente implica el uso de técnicas de modelado estadístico y matemático para extraer temas, temas o conceptos principales de un corpus de documentos (más diverso sea el conjunto de documentos que tenga, más temas o conceptos puedes generar). Hay modelos estadísticos probabilísticos, que utilizan técnicas estadísticas específicas que incluyen la descomposición de valores singulares y el LDA para descubrir estructuras semánticas latentes conectadas en datos de texto que generan temas y conceptos. Se utilizan ampliamente en análisis de texto e incluso en bioinformática.

3. **El resumen automatizado de documentos** es el proceso de utilizar un programa informático o algoritmo basado en técnicas estadísticas y de aprendizaje automático para resumir un documento o corpus de documentos. Existe una amplia variedad de técnicas para crear resúmenes de documentos automatizados, incluidas varias técnicas basadas en extracción y abstracción.

El concepto clave detrás de todos estos algoritmos es encontrar un subconjunto representativo del conjunto de datos original de modo que la esencia central del conjunto de datos desde los puntos de vista semántico y conceptual esté contenida en este subconjunto. Esto es un concepto recurrente que se puede aplicar a imagenes y texto

# Conceptos que debes conocer antes de empezar

 
 
1. **Documentos**:  una entidad que contiene texto y metadatos. Un corpus generalmente consiste en una colección de documentos. Estos documentos pueden ser oraciones simples o párrafos completos de información textual. El corpus tokenizado se refiere a un corpus en el que cada documento se tokeniza o se divide en tokens, que generalmente son palabras.

2. **Normalización de texto** proceso de limpieza, normalización y estandarización de datos textuales con técnicas como la eliminación de símbolos y caracteres especiales, la eliminación de etiquetas HTML extrañas, la eliminación de palabras vacías, la corrección ortográfica, stemming y la lematización.

3. **Feature extraction** es un proceso mediante el cual extraemos características o atributos significativos de datos textuales sin procesar para introducirlos en un algoritmo estadístico o de aprendizaje automático. Este proceso también se conoce como vectorización porque, por lo general, la transformación final de este proceso son vectores numéricos a partir de tokens de texto sin procesar. La razón es que los algoritmos convencionales funcionan con vectores numéricos y no pueden funcionar directamente con datos de texto sin procesar. Tenemos distintos mecanismos de abstraccion como BoW, TFIDF, Word Embeddings (**No te preocupes mas adelante vamos a ver mas a detalle esto)

4. **Matriz de diseño** se refiere a un mapeo de una colección de documentos a características donde cada fila indica un documento y cada columna indica una característica particular, generalmente una palabra o un conjunto de palabras. 

5. **Descomposicion SVD** La descomposición en valores singulares (SVD) es una técnica del álgebra lineal que se utiliza con bastante frecuencia en los algoritmos de resumen. SVD es el proceso de factorización de una matriz que es real o compleja. 



Los valores singulares de $\Sigma$ son particularmente importantes en los algoritmos de resumen ahora te preguntaras porque?. 

Usaremos SVD particularmente para la aproximación de matrices de bajo rango donde aproximamos la matriz original M con una matriz M tal que esta nueva matriz es una versión truncada de la matriz original M con un rango k y puede ser representada por SVD como $M = U \Sigma V^*$ 

Donde $\Sigma$ es una versión truncada de la matriz $\Sigma$  original, que ahora consiste solo en los k valores singulares más grandes, y los otros valores singulares están representados por cero.
