# Linguistica
La lingstica se define como el estudio cientfico del lenguaje, incluida la forma y la sintaxis del lenguaje, el significado y la semntica representados por el uso del lenguaje y el contexto de uso.

Aunque no se necesita una exploracin detallada de la lingstica para el anlisis de texto, es til conocer las diferentes reas de la lingstica porque algunas de ellas se utilizan ampliamente en el procesamiento del lenguaje natural y los algoritmos de anlisis de texto. Las principales reas distintivas de estudio en lingstica son las siguientes:

1. **Fonetica:** Este es el estudio de las propiedades acsticas de los sonidos producidos por el tracto vocal humano durante el habla.
2. **Fonologa:** Este es el estudio de los patrones de sonido interpretados en la mente humana y utilizados para distinguir entre diferentes fonemas para descubrir cules son significativos.
3. **Sintaxis:** Este suele ser el estudio de oraciones, frases, palabras y sus estructuras. Incluye investigar cmo las palabras se combinan gramaticalmente para formar frases y oraciones.
4. **Semantica:** el estudio del significado en el lenguaje y se puede subdividir en semntica lxica (significado de las palabras) y composicional (relaciones entre palabras).
5. ** Morfologia:** La morfologa es el estudio de la estructura y el significado de estas unidades distintivas o morfemas en un idioma. Un morfema es la unidad basica del lenguaje con un significado distintivo
6. **Lexico:** Este es el estudio de las propiedades de las palabras y frases utilizadas en un idioma y cmo construyen el vocabulario del idioma.
7. **Pragmatica:** Este es el estudio de cmo los factores lingsticos y no lingsticos, como el contexto y el escenario, pueden afectar el significado de una expresin de un mensaje o una declaracin.
8. **Analisis de discurso:** Este analiza el lenguaje y el intercambio de informacin en forma de oraciones a travs de conversaciones entre seres humanos.
9. **Estilistica:** Este es el estudio del idioma con un enfoque en el estilo de escritura, incluido el tono, el acento, el dilogo, la gramtica y el tipo de voz.
10. **Semiotica:** Este es el estudio de signos, smbolos y procesos de signos y cmo comunican significado.

# Sintaxis del lenguaje y estructura 

Conjunto de reglas, convenciones y principios especficos que generalmente gobiernan la forma en que las palabras se combinan en frases, las frases se combinan en clusulas y las clusulas se combinan en oraciones

1. **Palabras:** Las palabras son las unidades ms pequeas en un idioma que son independientes y tienen un significado propio. Pueden ser pronombres, verbos, adjetivos, pronombres, proposiciones, conjunciones, determinantes y advervios entre otros posibles tags
2. **Frases:** grupos de palabras forman frases, que forman el tercer nivel en el rbol sintctico.
3. **Clausulas:** Una clusula es un grupo de palabras con alguna relacin entre ellas que generalmente contiene un sujeto y un predicado (pueden ser declarativas, imperativas, relativas, interrogativas, exclamativas)

# Corpus
**Corpus** Conjunto de palabras dentro de un contexto determinado 
**Corpora** Los corpus de texto son colecciones grandes y estructuradas de textos o datos textuales, que generalmente consisten en cuerpos de texto escrito o hablado, a menudo almacenados en formato electrnico.

Los corpus de texto se crean con metadatos enriquecidos, lo que es extremadamente til para obtener informacin valiosa al utilizar los corpus para NLP y anlisis de texto. Las anotaciones populares para corpus de texto incluyen etiquetas de partes del discurso (POS), races de palabras, lemas y muchas ms. Estos son algunos de los mtodos y tcnicas ms utilizados para anotar corpus de texto:

1. POS Tagging
2. Word Stems
3. Word Lemmas
4. Constitucion de gramatica
5. Tipos de semantica y roles

Los Corpora mas comunes son:
1. Key Word in Context: KWIC creada en 1860 y se uso hasta 1950
2. Brown Corpus: un corpues de cerca de un millon de palabras en ingles creado en 1961
3. LOB Corpus: Lancaster-Oslo-Bergen (LOB) creado en 1970 creado con distintas fuentes
4. Collins Corpus: COBUILD creado en 1980 
5. CHILDES: Child Languaje Data Exchange System creado en 1984 con audios, vidoes y 26 lenguajes
6. WordNet: es una base de datos semantico-lexica orientado para el Ingles creado en 1985 por George Armitage en Princenton
7. Penn Treebank: creado en la universidad de Pensilvania 
8. BNC: British National Corpus tiene cerca de 100 millones de palabras del ingles creado en los 90
9. ANC: American NAtional Corpus es un corpus de 22 millones de palabras escritas y habladas desde 1990
10. COCA Corpus of Contemporary American English con 450 millones de palabras del ingles
11. Google N-gram Corpus: cerca de un trillon de palabras de fuentes de libros, paginas web etc creado en 2013
12. Reuters Corpus: coleccion de articulos e historias lanzado en los 2000

```python
import nltk # Esto puede demorar un rato
nltk.download('all')
```

```python
# Accedamos a un corpus e.g Brown Corpus
from nltk.corpus import brown
print('Categorias totales:', len(brown.categories()))
```

```python
print(brown.categories())
```

```python
# oraciones tokenizadas
brown.sents(categories='mystery')
```

```python
# Oraciones con POS tagging
brown.tagged_sents(categories='mystery')
```

```python
# Obtener oraciones en forma natural
oraciones = brown.sents(categories='mystery')
[' '.join(sentence_token) for sentence_token in oraciones]
```

```python
# Tages words
tagged_words = brown.tagged_words(categories='mystery')
# Obtener solo los nouns
nouns = [(word, tag) for word, tag in tagged_words if any(noun_tag in tag for noun_tag in ['NP', 'NN'])]
nouns
```

```python
nouns_freq = nltk.FreqDist([word for word, tag in nouns])
nouns_freq
```

```python
print(nouns_freq.most_common(10))
```

```python
# Un Corpus en espaol
from nltk.corpus import cess_esp
cess_esp = cess_esp.tagged_sents()
cess_esp
```

```python
# la etiqueta de 3 pers plur del pret. perf simpl es: vmis3p0
verbos_en_pasado = []
total = 0

for oracion in cess_esp:
    total += len(oracion)
    for item in oracion:
        if item[1] == 'vmis3p0':
            verbos_en_pasado.append(item[0])
        
print('Tengo', len(verbos_en_pasado), 'verbos en pasado y', total, 'palabras en total') 
print('El porcentaje de verbos es', len(verbos_en_pasado)/total)
```

```python
# Wordnet Corpus
from nltk.corpus import wordnet as wn
word = 'bike'
# Synsets
word_synsets = wn.synsets(word)
print(word_synsets)
```

```python
for synset in word_synsets:
    print('Synset Name:', synset.name())
    print('POS Tag:', synset.pos())
    print('Definicion:', synset.definition())
    print('Ejemplos:', synset.examples())
```
